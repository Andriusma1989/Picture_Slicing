# First slicing


```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load the image into a NumPy array
img = np.array(Image.open("/content/lenna1.jpg"))

# Define the number of slices
num_slices = 40

# Split the image into vertical strips
slices = np.array_split(img, num_slices, axis=1)

# Create two new arrays to hold the shredded image
shredded_img1 = np.zeros_like(img)
shredded_img2 = np.zeros_like(img)  

# Loop through the slices and add them to the shredded arrays one by one
pos1 = pos2 = 0
for i, slice in enumerate(slices):
    if i % 2 == 0:
        shredded_img1[:, pos1:pos1+slice.shape[1], :] = slice
        pos1 += slice.shape[1]
    else:
        shredded_img2[:, pos2:pos2+slice.shape[1], :] = slice
        pos2 += slice.shape[1] 

# Slicing the new images to remove unused space
shredded_img1 = shredded_img1[:, :pos1, :]
shredded_img2 = shredded_img2[:, :pos2, :]

#merge the 2 shreded image and rotate
concat_image = np.concatenate((shredded_img1,shredded_img2), axis=1)
first_slice = np.rot90(concat_image)

#display the original and sliced image 
plt.imshow(img)
plt.show()
plt.imshow(first_slice)
plt.show()

```
